```python
import numpy as np
import scipy.stats as S
import matplotlib.pyplot as plt
import math
```

```python
dicethrows = np.random.randint( low=1,high=7,size=(1000,2))
```

```python
samplemeans = np.mean(dicethrows,axis=1)
```

```python
print(samplemeans)
```

```python
markings = list(np.unique(samplemeans))
```

```python
markings
```

```python
minvalue = min(markings)
maxvalue = max(markings)
```

```python
binwidth = (markings[1] - markings[0])
```

```python
bins = len(markings)
```

```python
samplemeans = list(samplemeans)
```

```python
frequency = list(np.zeros((len(markings),)))
```

```python
for value in samplemeans:
    
    index = math.floor((value-minvalue)/binwidth)
    
    print(index)
    
    frequency[index] += 1
```

```python
print(frequency)
```

```python
plt.bar(markings,frequency)
```

```python
plt.show()
```

Now, let's talk about another way to plot Sampling frequency distributions.

Lets again throw 2 dices 1000 times.

```python
dicethrows = np.random.randint(low=1,high=7,size=(1000,2))
```

```python
samplemeans = np.mean(dicethrows,axis=1)
```

```python
print(samplemeans)
```

```python
markings = np.unique(samplemeans)
```

```python
print(markings)
```

```python
markings = list(markings)
```

```python
frequency = []
```

```python
for marking in markings:
    
    booleanmask = (samplemeans==marking)
    
    frequency.append(np.count_nonzero(samplemeans[booleanmask]))
```

```python
frequency
```

```python
plt.bar(markings,frequency)
```

```python
plt.show()
```
